```
#@title
from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.express as px
import matplotlib.cm as cm

import os
from skimage import io
from skimage.color import rgb2gray
from skimage import img_as_ubyte

import numpy as np
```

```
#@title
def show_image(_img,title_str=None):
  fig = px.imshow(_img, title=title_str, range_color=[0,255],color_continuous_scale='gray')
  fig.show()
```

```
#@title
def plot_image_histogram(_imgV,title_str=None):
  counts, bins = np.histogram(_imgV, range(257))
  fig = px.bar(bins[:-1] - 0.5, y=counts,range_x=[-0.5,255.5],labels={'x':'Intensity', 'y':'Count'},title=title_str)
  fig.update_layout(hovermode='x') # shows x-axis value also
  fig.show()
```

```
#@title
#img = io.imread('https://i.stack.imgur.com/04OgX.jpg')

img = io.imread('https://i.imgur.com/aLoSZWz.png')


if len(img.shape) == 3:
  img = rgb2gray(img)

img = img_as_ubyte(img, force_copy=False)

show_image(img,'Input Image')

imgV = img.ravel()
plot_image_histogram(imgV,'Input Image')
```

```
#@title
# Contrast stretching
imgV = imgV.astype(float)
min_g = np.min(imgV)
max_g = np.max(imgV)

new_imgV = np.ubyte(np.floor(255.0*(imgV - min_g)/(max_g - min_g)))
img_con = new_imgV.reshape((img.shape))

show_image(img_con,'After Contrast Enhancement')

imgV_con = img_con.ravel()
plot_image_histogram(imgV_con,'After Contrast Enhancement')

```
